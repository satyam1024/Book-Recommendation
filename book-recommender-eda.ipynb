# Data Exploration 

```python
books=pd.read_csv("/kaggle/input/7k-books-with-metadata/books.csv")
```

```python
books.head()
```

```python
books.info()
```

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.set(style="whitegrid")

print("📌 Shape of dataset:", books.shape)
print("\n📌 Data types and non-null values:")
print(books.info())
print("\n📌 Descriptive statistics:")
print(books.describe(include="all"))




```

```python
missing_count = books.isnull().sum()
missing_percent = (missing_count / len(books)) * 100
missing_df = pd.DataFrame({
    'Missing Values': missing_count,
    'Percentage (%)': missing_percent
}).sort_values(by='Missing Values', ascending=False)

print("\n📌 Missing values (count and %):")
print(missing_df)

plt.figure(figsize=(10, 6))
sns.heatmap(books.isnull().transpose(), cbar=False, cmap="viridis")
plt.xlabel("Columns")
plt.ylabel("Missing values")
plt.title("Missing Values Heatmap")
plt.show()



```

```python

categorical = books.select_dtypes(include="object")
for col in categorical.columns:
    print(f"\n📝 Top values in '{col}':\n", books[col].value_counts().head(10))


numerical = books.select_dtypes(include=["int64", "float64"])


books[numerical.columns].hist(bins=30, figsize=(15, 10), color="skyblue", edgecolor="black")
plt.suptitle("Histograms of Numerical Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2024 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm",
                      cbar_kws={"label": "Spearman correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

```python
book_missing = books.dropna(subset=["description", "num_pages", "average_rating", "published_year"])
```

```python
book_missing.head()
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(book_missing["words_in_description"], bins=100, kde=True, color="skyblue")
plt.title("Distribution of Words in Description")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")
plt.show()
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
book_missing_25_words.head()
```

```python

book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
book_missing_25_words.head()
```

```python
(
    book_missing_25_words
    .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index = False)
)
```

```python
sns.scatterplot(data=books, x="num_pages", y="average_rating")
plt.title("Pages vs Rating")
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x="published_year", y="average_rating", data=books)
plt.xticks(rotation=90)
plt.title("Published Year vs Average Rating")
plt.show()
```

```python
top_categories = books['categories'].dropna().str.split(", ").explode().value_counts().head(15)
top_authors = books['authors'].dropna().str.split(", ").explode().value_counts().head(15)

plt.figure(figsize=(12, 5))
sns.barplot(x=top_categories.values, y=top_categories.index, palette="viridis")
plt.title("Top 15 Book Categories")
plt.xlabel("Count")
plt.show()

plt.figure(figsize=(12, 5))
sns.barplot(x=top_authors.values, y=top_authors.index, palette="magma")
plt.title("Top 15 Authors")
plt.xlabel("Count")
plt.show()
```
